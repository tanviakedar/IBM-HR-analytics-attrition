```python
sns.countplot(x='MaritalStatus', hue='Attrition', data=df, palette='coolwarm')
plt.title('Marital Status vs Attrition')
plt.show()
# Percentage of attrition by marital status
print(df.groupby('MaritalStatus')['Attrition'].value_counts(normalize=True).unstack() * 100)
```

```python
# Singles have the highest attrition rate (the "Yes" bar is comparatively taller).
# Married employees make up the largest group, but their attrition rate is lower relative to their count.
# Divorced employees have the lowest attrition count.
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='JobRole', hue='Attrition', data=df, palette='Set3')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#1. Highest Attrition in Sales Executive & Laboratory Technician Roles
#Both roles show high counts of employees who left compared to others.
#Sales Executive and Laboratory Technician roles also have large total headcounts overall.

# 2.Research Scientist Also Shows Noticeable Attrition
#This role has a significant number of employees leaving, though not as much as the two roles above.

#3.Lowest Attrition in Manager and Healthcare Representative Roles
#Very few instances of attrition in these categories, suggesting better retention.

#4.Sales Representative Has a Relatively High Proportion of Attrition
# Despite having fewer total employees, the attrition proportion is high compared to others.

#5.Manufacturing Director, Research Director, and Human Resources Have Low Attrition
#These roles exhibit strong retention rates, with very few employees leaving.
```

```python
sns.countplot(x='Department', hue='Attrition', data=df, palette='Paired')
plt.title('Attrition by Department')
plt.show()
```

```python
# Research & Development (R&D) has the highest overall employee count: Attrition (leaving employees) is relatively lower compared to the total number staying.
# Sales Department shows a noticeable attrition rate:The gap between employees staying and leaving is smaller compared to R&D.
#                                                    Attrition is comparatively higher in Sales relative to its size.
# Human Resources has the lowest headcount overall:Attrition in HR is minimal, but given the small department size, even a few resignations could be impactful proportionally.
```

```python
sns.countplot(x='OverTime', hue='Attrition', data=df, palette={'Yes': 'tomato', 'No': 'mediumseagreen'})
plt.title('Attrition by Overtime')
plt.show()
```

```python
# Employees who work overtime have a higher attrition rate compared to those who don't.
# The number of employees leaving is significantly higher among those who are required to work overtime.
# Employees not working overtime show much lower attrition relative to their overall group size.
# Majority of employees without overtime responsibilities stay with the organization.
```

```python
sns.kdeplot(data=df[df['Attrition']=='Yes']['YearsAtCompany'], label='Left', shade=True)
sns.kdeplot(data=df[df['Attrition']=='No']['YearsAtCompany'], label='Stayed', shade=True)
plt.title('Years at Company Distribution')
plt.legend()
plt.show()
```

```python
# Employees who left the company (blue line) tend to have shorter tenures, mostly between 0 to 5 years.
# Employees who stayed (orange line) show a broader spread, with density peaking a bit later compared to those who left.
# After around 5â€“10 years, the attrition rate sharply drops, and most employees who stay tend to remain much longer.
# Very few employees leave after crossing around 10+ years at the company.
```

```python
sns.barplot(x='Attrition', y='MonthlyIncome', data=df, palette={'Yes': 'olive', 'No': 'magenta'},ci=None)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
# Employees who left ("Yes") have a lower average monthly income (~4800).
# Employees who stayed ("No") have a higher average monthly income (~6800).
# The income gap is significant â€” almost 2000 units difference.
```

```python
satisfaction_columns = ['JobSatisfaction', 'WorkLifeBalance', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']

# Group by Attrition and calculate the mean for satisfaction scores
mean_satisfaction = df.groupby('Attrition')[satisfaction_columns].mean()
print(mean_satisfaction)
```

```python
# Job and environment satisfaction show the largest drops (about -0.31) for those who left.
# Even relationship satisfaction and work-life balance are slightly lower among employees who left.
# This confirms that lower satisfaction across the board correlates with attrition.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='TotalWorkingHours', data=df)
plt.title('Attrition by Total Working hours')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='TotalWorkingYears', data=df, hue='Attrition',palette={'Yes':'magenta','No':'green'}, dodge=False)
plt.title('Attrition by Total Working Years')
plt.legend([],[], frameon=False)  
plt.show()
```

```python
# Attrition = Yes (Left):Median total working years is lower.
# Attrition = No (Stayed): Median total working years is higher.
# Interpretation: Employees who leave the company typically have less experience compared to those who stay longer.
# Attrition = Yes:Narrower IQR (middle 50% of data is closer together).
# Attrition = No:Wider IQR (more variation in total working years among those who stay).
# Interpretation:Employees who leave tend to have a more similar (low) amount of total working years, while those who stay have more diverse career lengths.
# Employees with fewer total working years are more likely to leave.
# Employees who stay tend to have longer careers.
# Attrition is higher among less experienced employees.
# There are a few outliers where even highly experienced employees leave, but itâ€™s rare.
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Attrition', y='MonthlyRate', data=df)
plt.title('Attrition by Monthly rate')
plt.show()
```
